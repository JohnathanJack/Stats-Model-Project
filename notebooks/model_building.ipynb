```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import statsmodels.api as sm
```

```python
complete_df = pd.read_csv('C:/Users/John/Documents/LHL Lecture Material/Stats-Model-Project/data/complete_df.csv')
```

```python
df_removed_limit = complete_df[(complete_df['yelp_num_res'] != 50) & (complete_df['num_of_res'] != 50)]
df_removed_limit
```

```python
fsquare_removed = df_removed_limit[(df_removed_limit['yelp_rating'].notna())].drop(['name','address','coordinates','rating','total_photos','total_ratings','num_of_res'], axis=1)
```

```python
y = fsquare_removed['num_of_bikes']
x_yelp = fsquare_removed.drop('num_of_bikes',axis=1)
x_yelp = sm.add_constant(x_yelp)
```

Provide model output and an interpretation of the results. 

```python
model = sm.OLS(y, x_yelp)
results = model.fit()
print(results.summary())
```
